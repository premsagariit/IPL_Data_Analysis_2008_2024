```python
match_data["target_overs"].fillna(match_data["target_overs"].median(), inplace=True)
match_data["target_runs"].fillna(match_data["target_runs"].median(), inplace=True)
```

```python
match_data["player_of_match"].fillna('unknown', inplace=True)
```

There is insifficient data collected about the city in about `51` matches. Filling thises ceels with the string `missing`

## 1. Points Table of IPL Season

```python
def points_table_season(season):
    season_df = match_data[(match_data["season"] == season) & (match_data["match_type"] == "League")]
    return season_df

# Define a function to calculate the points table for a season
def calculate_points_table(season_df):
    team_stats = np.sort(season_df['team1'].unique())
    
    points_table = pd.DataFrame()
    points_table["Team Name"] = team_stats
    points_table["Total Matches"] = 14
    
    wins = season_df["winner"].value_counts().reindex(team_stats, fill_value=0).sort_index().values
    losses = season_df["looser"].value_counts().reindex(team_stats, fill_value=0).sort_index().values
    
    points_table["Wins"] = wins
    points_table["Loss"] = losses
    points_table["Points"] = points_table["Wins"] * 2

    # Since there are a few cancelled matches at the League stage, for those mathces, the points are equally distributed to both the team_stats i.e., 1 point each
    for index, row in points_table.iterrows():
        if row["Wins"] + row["Loss"] != 14:
            points_table.at[index, "Points"] += (14 - (row["Wins"] + row["Loss"]))
    points_table.sort_values("Points", ascending=False).reset_index(drop=True)

    for index, row in points_table.iterrows():
        if row["Wins"] + row["Loss"] != 14:
            points_table.at[index, "Points"] += (14 - (row["Wins"] + row["Loss"]))
    points_table = points_table.sort_values("Points", ascending=False).reset_index(drop=True)
    
    return points_table

def display_points_table(season):
    season_df = points_table_season(season)
    points_table = calculate_points_table(season_df)
    display(points_table)

season_dropdown = widgets.Dropdown(
    options=sorted(match_data["season"].unique()),  # Unique season numbers
    value=17,  # Default value
    description='Season:',
)

widgets.interactive(display_points_table, season=season_dropdown)
```

## 5. Season Stats

```python
# Concate the two dataframes
complete_df = pd.merge(match_data, deliveries_data, how='inner', on="match_id")
complete_df.head()
```

```python
# Create the Season DataFrame
season_stats = pd.DataFrame()
season_stats["Season"] = complete_df["season"].unique()
season_stats.head();
```

```python
# Number of matches played in each season
season_stats["Total Matches Played"] = season_stats["Season"].apply(lambda x: match_data[match_data["season"]==x].shape[0])
season_stats.head()
```

```python
# Total runs in each season
season_stats["Total Runs"] = season_stats["Season"].apply(lambda x: complete_df[complete_df["season"]==x]["total_runs"].sum())
season_stats.head()
```

```python
# Average number of runs per macth in each season
season_stats["Average Runs"] = (season_stats["Total Runs"]/season_stats["Total Matches Played"]).round(2)
season_stats.head()
```

```python
# Number balls bowled and nuber of overs bowled
season_stats["Total Balls"] = season_stats["Season"].apply(lambda x: complete_df[complete_df["season"]==x]["ball"].count())
season_stats["Total Overs"] = (season_stats["Total Balls"]/6).round()
season_stats.head()
```

```python
# Number of Wickets and Average number of wickets in each season 
season_stats["Total Wickets"] = season_stats["Season"].apply(lambda x: complete_df[complete_df["season"]==x]["is_wicket"].sum())
season_stats["Average Wickets"] = (season_stats["Total Wickets"]/season_stats["Total Matches Played"]).round()
season_stats.head()
```

```python
# Winners and Runners of each season
finals_df = match_data[match_data["match_type"]=="Final"].reset_index(drop=True)
season_stats["Winners"] = finals_df["winner"]
season_stats["Runners"] = finals_df["looser"]
season_stats.head()
```

```python
# Most runs scored by the bastmen in each season
season_batter_runs = complete_df.groupby(["season", "batter"])["batsman_runs"].sum().reset_index()
season_highest_batter_runs = season_batter_runs.loc[season_batter_runs.groupby("season")["batsman_runs"].idxmax()].reset_index(drop=True)
season_stats["Most Runs Scorer"] = season_highest_batter_runs["batter"].astype(str) + "-" + season_highest_batter_runs["batsman_runs"].astype(str)
season_stats.head()
```

```python
# Most Wickets taken by the Bowler
season_bowler_wickets = complete_df.groupby(["season", "bowler"])["is_wicket"].sum().reset_index()
season_highest_bowler_wickets = season_bowler_wickets.loc[season_bowler_wickets.groupby("season")["is_wicket"].idxmax()].reset_index(drop=True)
season_stats["Most Wickets Taken"] = season_highest_bowler_wickets["bowler"].astype(str) + "-" + season_highest_bowler_wickets["is_wicket"].astype(str)
season_stats.head()
```

## 3. Team Stats

```python
# create the team stats DataFrame
team_stats = pd.DataFrame()
team_stats["Team Name"] = match_data['winner'].unique()
team_stats = team_stats.sort_values(by="Team Name").reset_index(drop=True)
```

```python
# Number of matches played by each team
team_match = match_data["team1"].value_counts()+match_data["team2"].value_counts()
team_stats["Total Matches Played"] = team_match.values
team_stats.head()
```

```python
# Number of times each team won the IPL Season Cup
team_stats["Season Winners"] = team_stats["Team Name"].apply(lambda x: finals_df[finals_df["winner"]==x]["winner"].count())
team_stats = team_stats.sort_values(by="Season Winners", ascending = False).reset_index(drop=True)
team_stats.head()
```

```python
# Number of matches won by each team
team_stats["Won"] = team_stats["Team Name"].apply(lambda x: len(match_data[match_data['winner']==x]))
team_stats.head()
```

```python
# Number of Matches lost by each team
team_stats["Lost"] = team_stats["Team Name"].apply(lambda x: len(match_data[match_data["looser"]==x]))
team_stats.head()
```

```python
# Win percentage of each team
team_stats["Win %"] = (team_stats["Won"]/team_stats["Total Matches Played"]*100).round(2)
team_stats.head()
```

```python
# Loss percentage of each team
team_stats["Loss %"] = (team_stats["Lost"]/team_stats["Total Matches Played"]*100).round(2)
team_stats.head()
```

```python
# Total Runs scored by each team in all the matches played till 2024
team_stats["Total Runs"] = team_stats["Team Name"].apply(lambda x: deliveries_data[deliveries_data["batting_team"]==x]["total_runs"].sum())
team_stats.head()
```

```python
# Total Wickets taken by each team in all the matches played till 2024
team_stats["Total Wickets"] = team_stats["Team Name"].apply(lambda x: deliveries_data[deliveries_data["bowling_team"]==x]["is_wicket"].sum())
team_stats.head()
```

```python
# Number of balls Batted by each team in all the seasons
team_stats["Total Balls Played"] = team_stats["Team Name"].apply(lambda x: deliveries_data[deliveries_data["batting_team"]==x].shape[0])
team_stats.head()
```

```python
# Number of overs bowled and Run Rate of each team
team_stats["Total Overs"] = (team_stats["Total Balls Played"]/6).round()
team_stats["Run Rate"] = (team_stats["Total Runs"]/(team_stats["Total Overs"])).round(3)
team_stats.head()
```

```python
# Average runs scored by each team
team_stats["Average Runs"] = team_stats["Total Runs"]/team_stats["Total Matches Played"]
team_stats.head()
```

```python
# Average wicktes taken
team_stats["Average Wickets"] = (team_stats["Total Wickets"]/team_stats["Total Matches Played"]).round()
team_stats.head().head()
```

```python
# Economy Rate of each team
team_stats["Economy Rate"] = (team_stats["Total Runs"]/team_stats["Total Overs"]).round(3)
team_stats.head()
```

```python
# Matches that has no result are not considered for count of 4 and 6 
no_result_match_id = [501265, 829763, 829813, 1178424, 1359519]
cleaned_deliveries = deliveries_data[~deliveries_data['match_id'].isin(no_result_match_id)]
all_4= cleaned_deliveries[cleaned_deliveries["batsman_runs"]==4]
all_6 = cleaned_deliveries[cleaned_deliveries["batsman_runs"]==6]
```

```python
team_stats["6's"] = team_stats["Team Name"].apply(lambda x: all_6[all_6["batting_team"]==x].shape[0])
team_stats["4's"] = team_stats["Team Name"].apply(lambda x: all_4[all_4["batting_team"]==x].shape[0])
team_stats.head()
```

```python
scores = cleaned_deliveries.groupby(["match_id", "batting_team"])["total_runs"].sum().reset_index()
```

```python
# Highest and Lowest Scores of each team
team_stats["Highest Score"] = team_stats["Team Name"].apply(lambda x: scores[scores["batting_team"]==x]["total_runs"].max())
team_stats["Lowest Score"] = team_stats["Team Name"].apply(lambda x: scores[scores["batting_team"]==x]["total_runs"].min())
team_stats.head()
```

```python
# Filter out Powerplay and Death overs
powerplay_overs = deliveries_data[deliveries_data["over"]<=6]
death_overs = deliveries_data[deliveries_data["over"]>15]
```

```python
# Total Powerplay score, Total Death over scores, Average Powerplay score, Average Death Over scores of each team
team_stats["Powerplay Score"] = team_stats["Team Name"].apply(lambda x: powerplay_overs[powerplay_overs["batting_team"]==x]["total_runs"].sum())
team_stats["Average Powerplay Score"] = (team_stats["Powerplay Score"]/team_stats["Total Matches Played"])
team_stats["Death Over Scores"] = team_stats["Team Name"].apply(lambda x: death_overs[death_overs["batting_team"]==x]["total_runs"].sum())
team_stats["Average Death Over Scores"] = team_stats["Death Over Scores"]/team_stats["Total Matches Played"]
team_stats.head()
```

```python
# Create DataFrame called Batter Stats
batter_stats = pd.DataFrame()

# Filter out the balls with no extras
no_extras = deliveries_data[deliveries_data["extra_runs"]==0]
batters = no_extras["batter"].value_counts().reset_index()
```

```python
batter_stats["Player Name"] = batters["batter"]
batter_stats.head()
```

```python
# Total runs scored by the batter
batter_stats["Total Runs"] = batter_stats["Player Name"].apply(lambda x: no_extras[no_extras["batter"]==x]["batsman_runs"].sum())
batter_stats = batter_stats.sort_values(by ="Total Runs", ascending=False).reset_index(drop=True).head(10)
batter_stats.head()
```

```python
# Number of balls played by the batter
batter_stats["Balls Faced"] = batter_stats["Player Name"].apply(lambda x: no_extras[no_extras["batter"]==x]["ball"].count())
batter_stats.head()
```

```python
# Number of matches played by the batter
unique_match =deliveries_data.groupby(["match_id", "batter"])["batter"].value_counts().reset_index()
batter_stats["No.of Matches Played"] = batter_stats["Player Name"].apply(lambda x:unique_match[unique_match["batter"]==x]["count"].count())
batter_stats.head()
```

```python
# Average runs of the batter
batter_stats["Average Runs"] = (batter_stats["Total Runs"]/batter_stats["No.of Matches Played"]).round(2)
batter_stats.head()
```

```python
# Strike rate for the batter
batter_stats["Strike Rate"] = (batter_stats["Total Runs"]/batter_stats["Balls Faced"]*100).round(2)
batter_stats.head()
```

```python
# Number of 6's, 4's, 3's, 2's, 1's, 0's
balls_6 = no_extras[no_extras["batsman_runs"]==6]
balls_4 = no_extras[no_extras["batsman_runs"]==4]
balls_3 = no_extras[no_extras["batsman_runs"]==3]
balls_2 = no_extras[no_extras["batsman_runs"]==2]
balls_1 = no_extras[no_extras["batsman_runs"]==1]
balls_0 = no_extras[no_extras["batsman_runs"]==0]
batter_stats["6's"]=batter_stats["Player Name"].apply(lambda x: balls_6[balls_6["batter"]==x]["ball"].count())
batter_stats["4's"]=batter_stats["Player Name"].apply(lambda x: balls_4[balls_4["batter"]==x]["ball"].count())
batter_stats["3's"]=batter_stats["Player Name"].apply(lambda x: balls_3[balls_3["batter"]==x]["ball"].count())
batter_stats["2's"]=batter_stats["Player Name"].apply(lambda x: balls_2[balls_2["batter"]==x]["ball"].count())
batter_stats["1's"]=batter_stats["Player Name"].apply(lambda x: balls_1[balls_1["batter"]==x]["ball"].count())
batter_stats["0's"]=batter_stats["Player Name"].apply(lambda x: balls_0[balls_0["batter"]==x]["ball"].count())
batter_stats.head()
```

```python
# Number of times the batter got out
wickets = deliveries_data[deliveries_data["is_wicket"]!=0]
batter_stats["No.of Dismissals"] = batter_stats["Player Name"].apply(lambda x: wickets[wickets["batter"]==x]["is_wicket"].count())
batter_stats.head()
```

```python
# Number of Player of the match awards won by each player
batter_stats["No.of Player of Matches"] = batter_stats["Player Name"].apply(lambda x: match_data[match_data["player_of_match"]==x]["player_of_match"].count()) 
batter_stats.head()
```

```python
# Number of Ducks (0 runs in a match) by each batter 
duck_in_a_match = runs_in_a_match[runs_in_a_match["total_runs"]==0].reset_index()
batter_stats["Number of Ducks"] = batter_stats["Player Name"].apply(lambda x: duck_in_a_match[duck_in_a_match["batter"]==x].shape[0])
batter_stats.head()
```

## 5. Bowler Stats

```python
# Create the dataframe called bowler stats
bowler_stats = pd.DataFrame()
bowler_stats["Player Name"] = deliveries_data["bowler"].value_counts().reset_index().bowler
bowler_stats.head()
```

```python
# Number of wickets taken by bowler in ipl career
wickets = deliveries_data[deliveries_data["is_wicket"]!=0]
bowler_stats["Wickets Taken"] = bowler_stats["Player Name"].apply(lambda x: wickets[wickets["bowler"]==x]["is_wicket"].sum())
bowler_stats = bowler_stats.sort_values(by="Wickets Taken", ascending=False).reset_index(drop=True).head(10)
bowler_stats.head()
```

```python
# Number of Matches played
match_count_bowler_stats= deliveries_data.groupby(["match_id"])["bowler"].value_counts().reset_index()
bowler_stats["Total Matches Played"] = bowler_stats["Player Name"].apply(lambda x: match_count_bowler_stats[match_count_bowler_stats["bowler"]==x].shape[0])
bowler_stats.head()
```

```python
# Number of balls bowled
bowler_stats["Balls Bowled"] = bowler_stats["Player Name"].apply(lambda x: no_extras[no_extras["bowler"]==x].shape[0])
bowler_stats.head()
```

```python
# Number of runs given
bowler_stats["Runs Given"] = bowler_stats["Player Name"].apply(lambda x: deliveries_data[deliveries_data["bowler"]==x]["total_runs"].sum())
bowler_stats.head()
```

```python
# Number of overs bowled and the economy of the bowler
bowler_stats["Overs"] = (bowler_stats["Balls Bowled"]/6).round()
bowler_stats["Economy"] = (bowler_stats["Runs Given"]/bowler_stats["Overs"]).round(3)
bowler_stats.head()
```

```python
# Average runs given per match
bowler_stats["Average Runs Given"] = (bowler_stats["Runs Given"]/bowler_stats["Total Matches Played"]).round(2)
bowler_stats.head()
```

```python
# Bowlers best bowling figure
bowler_best = deliveries_data.groupby(["match_id", "bowler"]).agg(
    total_runs_given = ("total_runs", 'sum'),
    total_wickets = ("is_wicket", "sum")
).reset_index()

max_wickets = bowler_best.loc[bowler_best.groupby(["bowler"])["total_wickets"].idxmax()]
bowler_stats["Highest Wickets"] = bowler_stats["Player Name"].apply(lambda x: max_wickets[max_wickets["bowler"]==x]["total_wickets"].max())
bowler_stats["Runs"] = bowler_stats["Player Name"].apply(lambda x: max_wickets[max_wickets["bowler"]==x]["total_runs_given"].max())
bowler_stats["Best Performance"] = bowler_stats["Highest Wickets"].astype(str) + " - " + bowler_stats["Runs"].astype(str)
bowler_stats.drop(["Highest Wickets", "Runs"], axis=1, inplace=True)
bowler_stats.head()
```

```python
# Number of Maiden overs
over_summary = deliveries_data.groupby(["match_id", "bowler", "over"])["total_runs"].sum().reset_index()
maidem_over = over_summary[over_summary["total_runs"]==0]
bowler_stats["No.of Maidens"] = bowler_stats["Player Name"].apply(lambda x: maidem_over[maidem_over["bowler"]==x].shape[0])
bowler_stats.head()
```

```python
# Number os 6's, 4's, 3's, 2's, 1's, dots, extras given by bowler
all_6 = deliveries_data[deliveries_data["batsman_runs"]==6]
all_4 = deliveries_data[deliveries_data["batsman_runs"]==4]
all_3 = deliveries_data[deliveries_data["batsman_runs"]==3]
all_2 = deliveries_data[deliveries_data["batsman_runs"]==2]
all_1 = deliveries_data[deliveries_data["batsman_runs"]==1]
all_0 = deliveries_data[deliveries_data["total_runs"]==0]
extras = deliveries_data[deliveries_data["extra_runs"]!=0]
bowler_stats["6's"] = bowler_stats["Player Name"].apply(lambda x: all_6[all_6["bowler"]==x].shape[0])
bowler_stats["4's"] = bowler_stats["Player Name"].apply(lambda x: all_4[all_4["bowler"]==x].shape[0])
bowler_stats["3's"] = bowler_stats["Player Name"].apply(lambda x: all_3[all_3["bowler"]==x].shape[0])
bowler_stats["2's"] = bowler_stats["Player Name"].apply(lambda x: all_2[all_2["bowler"]==x].shape[0])
bowler_stats["1's"] = bowler_stats["Player Name"].apply(lambda x: all_1[all_1["bowler"]==x].shape[0])
bowler_stats["No.of Dots"] = bowler_stats["Player Name"].apply(lambda x: all_0[all_0["bowler"]==x].shape[0])
bowler_stats["Extras"] = bowler_stats["Player Name"].apply(lambda x: extras[extras["bowler"]==x]["total_runs"].sum())
bowler_stats.head()
```

```python
# Number of Player of the match awards won by each player
bowler_stats["No.of Player of Matches"] = bowler_stats["Player Name"].apply(lambda x: match_data[match_data["player_of_match"]==x]["player_of_match"].count()) 
bowler_stats.head()
```

```python
# Maximum runs given in a match
runs = deliveries_data[deliveries_data["total_runs"]!=0]
bowler_max_runs = runs.groupby(["match_id", "bowler"])["total_runs"].sum().reset_index()
bowler_stats["Max Runs Given"] = bowler_stats["Player Name"].apply(lambda x: bowler_max_runs[bowler_max_runs["bowler"]==x]["total_runs"].max())
bowler_stats.head()
```

## Hypothesis Testing

### 1. Toss Winning and Match Winning

**Hypothesis**:

Null Hypothesis: There is no relationship between winning the toss and winning the match

Alternate Hypothesis: There is significant relationship between winning the toss and winning the match

```python
toss_winner_match_winner = match_data["toss_winner"]==match_data["winner"]

contingency_table = pd.crosstab(match_data['toss_winner'], toss_winner_match_winner, rownames=["Toss Winner is Match Winner"], colnames=["Toss Winner"])
contingency_table
```

```python
# Perform the Chi - Square test
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)

print('Chi-square test statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
# print('Expected frequencies:', expected)
```

Checking for the 95% confidence level

$alpha$=0.05

```python
alpha = 0.05
if p<alpha:
    print("We reject the Null Hypothesis. There is a significant relationship between winning the toss and winning the match")
else:
    print("Fail to reject the null hypothesis. There is no significant relationship between winning the toss and winning the match.")
```

## Logistic Regression

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
df = match_data[['city', 'match_type', 'venue', 'player_of_match', 'team1', 'team2', 'toss_winner', 'toss_decision', 'target_runs', 'target_overs', 'result_margin', 'winner']]
```

```python
label_encoder = LabelEncoder()
columns_to_encode = list(df.columns)
df[columns_to_encode] = df[columns_to_encode].apply(lambda x: label_encoder.fit_transform(x))
```

```python
df.head()
```

```python
df['team1_win'] = (df['winner'] == df['team1']).astype(int)
```

```python
# Split data into X (features) and y (target)
X = df.drop(columns=['winner', 'team1_win'])
y = df['team1_win']
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

logreg_model = LogisticRegression(max_iter=500)
logreg_model.fit(X_train, y_train)
```

```python
logreg_model.score(X_test, y_test)
```

```python
y_preds = logreg_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_preds)
```

```python
print(classification_report(y_test, y_preds))
```
